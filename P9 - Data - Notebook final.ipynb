# 2. Jointures

## 2.1 : merge_1 = population + disponibilité alimentaire (df_1)

```python
merge_1 = pd.merge (population, df_1, on = 'Zone', how='outer', indicator = True)
```

```python
# On a vu plus haut que le fichier population a plus de pays (238) que le fichier disponibilité alimentaire (174)
# En résultat de cette jointure on devrait donc retrouver des "both" et des "left only"

merge_1._merge.unique()
```

```python
leftonly = merge_1.loc[merge_1['_merge'] == 'left_only', :]
```

```python
# On a bien 64 pays de différence entre les deux fichiers

leftonly.Zone.nunique()
```

```python
#On affiche les pays n'ayant pas de correspondance dans le fichier disponibilité alimentaire:
leftonly.Zone.unique()

#Ce sont principalement des iles qui sont souvent indexées à un autre pays = on décide de ne pas les conserver.
```

```python
# On ne retient que les pays qui ont une correspondance entre les deux fichiers:

merge_1 = merge_1.loc[merge_1['_merge'] == 'both']
```

```python
merge_1.drop('_merge', axis=1, inplace=True)
```

## 2.2 : merge_2 = pib + merge_1

```python
merge_2 = pd.merge (pib, merge_1, on = 'Zone', how='outer', indicator = True)
```

```python
merge_2._merge.unique()
```

```python
# Sélection des pays n'ayant que le PIB qui ressort mais pas de correspondance avec le fichier principal "merge_1":

merge_2.loc[merge_2['_merge'] == 'left_only', :]

# Ce sont également principalement des iles et territoires annexés à un autre pays "principal", on ne les retient donc pas.
```

```python
# Toutefois on remarque qu'il y a la Chine qui ressort
# Dans le fichier PIB il y a la Chine et la Chine continentale 
# Mais cela semble être les mêmes, on ne concervera donc que la Chine continentale

pib.loc[(pib['Zone'] == 'Chine') | (pib['Zone'] =='Chine, continentale') | (pib['Zone'] =='Chine - RAS de Macao') | (pib['Zone'] =='Chine - RAS de Hong-Kong'), :]
```

```python
# Pour les "right only" on ne trouve pas de correspondance au niveau du PIB, on décide de ne pas concervers Taiwan.

rightonly_1 = merge_2.loc[merge_2['_merge'] == 'right_only', :]
rightonly_1
```

```python
# On ne concerve que les pays qui trouvent une correspondance entre les deux fichiers

merge_2 = merge_2.loc[merge_2['_merge'] == 'both']
```

```python
merge_2.drop('_merge', axis=1, inplace=True)
```

## 2.3 : merge_3 = stabilité politique + merge_2

```python
merge_3 = pd.merge (politique, merge_2, on = 'Zone', how='outer', indicator = True)
```

```python
merge_3._merge.unique()
```

```python
# Pays qui ne trouvent pas de correspondance dans le fichier stabilité politique
merge_3.loc[merge_3['_merge'] == 'right_only', :]

# Même si on a pas sa stabilité politique on conserve quand même la Chine continentale car il s'agit d'une grande "région"
# On ne conserve pas les iles francaises puisque de toute façon nous recherchons à exporter à l'étranger.
```

```python
# Pays qui ressortent pour la stabilité politique mais qui n'ont pas de correspondance dans le fichier merge_2:
merge_3.loc[merge_3['_merge'] == 'left_only', :]

# Pour ces pays il nous manque toutes les informations importantes nécessaires à notre étude, il y a trop de données manquantes,
# On décide de ne pas les conserver.
```

```python
# On ne garde que les pays qui ont une correspondance entre les deux fichiers + la Chine continentale

merge_3 = merge_3.loc[(merge_3['_merge'] == 'both')| (merge_3['Zone'] == 'Chine, continentale')]
```

```python
merge_3.drop('_merge', axis=1, inplace=True)
```

# 2.4 : merge 4 = douanes + merge_3

```python
merge_4 = pd.merge (douanes, merge_3, on = 'Zone', how='outer', indicator = True)
```

```python
merge_4._merge.unique()
```

```python
# On affiche les pays qui ne ressortent que dans le fichier douanes et ne trouvent pas de correspondance dans le fichier principal

merge_4.loc[merge_4['_merge'] == 'left_only', :]

# Il s'agit des continents et quelques pays.
# On supprime toutes ces lignes puisque de toute facon il nous manque trop d'informations pour les conserver.
```

```python
# Pour ces pays là on n'a pas les frais de douanes, on décide tout de même de les conserver

merge_4.loc[merge_4['_merge'] == 'right_only', :]
```

```python
merge_4 = merge_4.loc[(merge_4['_merge'] == 'both')| (merge_4['_merge'] == 'right_only')]
```

```python
merge_4.drop('_merge', axis=1, inplace=True)
```

```python
df = merge_4
```

## 2.5 : df final 

```python
# On supprimer toutes les colonnes que l'on estime de trop, qui ne devrait pas nous servir.

df = df.drop(['Alimentation pour touristes', 
         'Aliments pour animaux', 
         'Autres utilisations (non alimentaire)',
        'Disponibilité alimentaire en quantité (kg/personne/an)',
        'Disponibilité de matière grasse en quantité (g/personne/jour)', 
        'Disponibilité de protéines en quantité (g/personne/jour)',
        'Pertes', 'Résidus', 'Semences', 'Traitement', 'Variation de stock', 'Nourriture', 'Disponibilité alimentaire (Kcal/personne/jour)'], axis=1)
```

```python
# Il faut que l'on étudie les NAN car le fichier ne peut pas en conserver:

df.info()
#la variable "exportations" est la seule comportant un taux de NAN important
```

```python
# On l'affiche:
df.loc[df["Exportations - Quantité"].isnull()]

# On voit deux pays pour lesquels on a aucune info sur les autres variables de dispo alimentaires: Ouzbékistan et
    #République démocratique populaire lao, on considère les infos insuffisantes: on supprime ces deux pays.
# Pour les Maldives et Djibouti on a pas non plus l'information de production, on décide donc de supprimer également
# Pour les autres pays on connait toutes les autres variables, on remarque aussi que leur importation + leur production = leur
    # disponibilité interieure, on peut donc se dire que peut etre ces pays là n'exporte pas du tout,on choisi de remplacer 
    # par 0.
```

```python
df.drop(df.loc[df['Zone'] == 'Ouzbékistan'].index, inplace=True)
```

```python
df.drop(df.loc[df['Zone'] == 'République démocratique populaire lao'].index, inplace=True)
```

```python
df.drop(df.loc[df['Zone'] == 'Maldives'].index, inplace=True)
```

```python
df.drop(df.loc[df['Zone'] == 'Djibouti'].index, inplace=True)
```

```python
df['Exportations - Quantité'] = df['Exportations - Quantité'].fillna(0)
```

```python
# Pour les restes des NAN puisque leur nombre est très faible on décide de remplacer par la médiane.
df = df.fillna(df.median())
```

```python
# On regarde globalement toutes nos variables, pour voir notamment si on a l'air d'avoir beaucooup d'outliers:
# Pour ce faire on regarde par exemple le min et le max de chaque variable
# La variable population risque de créer des incohérences avec l'ACP qui est très sensible aux outliers, on décidera de mettre
    #cette variable de côté puisque de toute façon la seule raison pour laquelle on l'utiliserait serait pour connaitre le
    # nombre d'habitant si à la fin cela devait être un critère décisif dans le choix des pays.
    
df.describe()
```

```python
# On modifie l'index « numéro de ligne » du df par la colonne "Zone" du df:
df.set_index('Zone')
```

```python
# Vérification qu'il n'y a plus de NAN:
df.info()
```

```python
df.to_csv('P9_df_final.csv', index=False)
```
